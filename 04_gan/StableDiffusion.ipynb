```python
from stable_diffusion_tf.stable_diffusion import Text2Image
from PIL import Image
import matplotlib.pyplot as plt
```

```python
generator = Text2Image(
    img_height=512,
    img_width=512,
    jit_compile=False,
)
```

```python
p = 'pine forest, brown bear, illustration, oil painting'
```

```python
img = generator.generate(
    p,
    num_steps=30,
    unconditional_guidance_scale=6,
    temperature=0.6,
    batch_size=1,
    seed=42
)
plt.imshow(img[0])
```

```python
Image.fromarray(img[0]).save("stable_diffusion.png")
```
