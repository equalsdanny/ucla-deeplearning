```python
from scipy.stats import truncnorm, norm
import numpy as np
import pandas as pd
import seaborn as sns
%config InlineBackend.figure_format = 'retina'
sns.set_context('talk')
```

```python
a = 5
x = np.linspace(-a, a, 1000)
curves = pd.DataFrame({
    'trunc. to [-1, 1]': truncnorm.pdf(x, -1, 1),
    'trunc. to [-0.5, 0.5]': truncnorm.pdf(x, -0.5, 0.5),
    'original normal': norm.pdf(x)
}, index=x)
```

```python
curves.plot(title='PDF of Normal Distribution (mean = 0, std = 1)', figsize=[10, 6])
```
